# Formación de senales básicas

Sea la siguiente señal continua:

$$
x_1(t) = \rho(t-1)-\rho(t-2)-u(t-1)
$$

a) Graficar $x_1(t)$ y su derivada.

b) Graficar $x_2 = x_1(t-2)$ y $x_3 = x_1(-2t+1)$.


```python
import numpy as np
import matplotlib.pyplot as plt


# Definición de la función escalón unitario
def u(t):
    return np.heaviside(t, 1)


# Definición de la función rampa
def r(t):
    return t * u(t)


# Definición del rango de tiempo
t = np.arange(-4, 8, 0.01)

# Definición de la señal x1(t)
x1 = r(t - 1) - r(t - 2) - u(t - 5)

# Cálculo de la derivada de la señal x1(t)
dx = t[1] - t[0]
x2 = np.diff(x1) / dx

# Gráfica de la señal original y su derivada
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, x1)
plt.grid(True)
plt.title("Funcion Original")
plt.subplot(2, 1, 2)
plt.plot(t[1:], x2)
plt.ylim([-2, 2])
plt.grid(True)
plt.title("Funcion Derivada")

# Ejemplo de desplazamiento e inversión de señales
tx = t - 2
x2 = r(tx - 1) - r(tx - 2) - u(tx - 5)
tx = -2 * t + 1
x3 = r(tx - 1) - r(tx - 2) - u(tx - 5)

# Gráfica de las señales desplazadas e invertidas
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, x2)
plt.grid(True)
plt.title("Funcion Desplazada 2")
plt.subplot(2, 1, 2)
plt.plot(t, x3)
plt.grid(True)
plt.title("Funcion x3")
plt.show()
```

# Señales básicas de tiempo continuo y de tiempo discreto

Sea $f(t) = \rho(t) - \rho(t-10) - 10 * u(t-10)$. Se pide gráficar con python:

> a) Señal de tiempo continuo: $f(t)$

> b) Señal de tiempo discreto: $f[n]$


```python
import numpy as np
import matplotlib.pyplot as plt


# Definición de la función escalón unitario
def U(n):
    return np.heaviside(n, 1)


# Definición de la función rampa
def R(n):
    return n * U(n)


# Definición del intervalo de tiempo continuo
dt = 0.01
t = np.arange(-5, 15 + dt, dt)

# Función de tiempo continuo f(t)
f = R(t) - R(t - 10) - 10 * U(t - 10)

# Gráfica de la función de tiempo continuo
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t, f)
plt.grid(True)
plt.title("Funcion de Tiempo Continuo")

# Definición del intervalo de tiempo discreto
n = np.arange(-5, 16)

# Función de tiempo discreto f[n]
f_discreta = R(n) - R(n - 10) - 10 * U(n - 10)

# Gráfica de la función de tiempo discreto
plt.subplot(2, 1, 2)
# plt.stem(n, f_discreta)
plt.stem(n, f_discreta, linefmt="b-", markerfmt="bo", basefmt=" ")
plt.grid(True)
plt.title("Funcion Discreta")
plt.tight_layout()
plt.show()
```

# Cálculo de energía

Dada las siguientes señales: $y_1(t) = e^{2t}u(t);y_2(t)=3(\frac{1}{2})^n u[n]$


```python
import numpy as np
import matplotlib.pyplot as plt


# Definición de la función escalón unitario
def U(t):
    return 1 * (t >= 0)


# Definición de la función exponencial de decrecimiento
def exp_decay(t):
    return np.exp(-2 * t) * U(t)


# Definición de la función exponencial de decrecimiento discreta
def exp_decay_discrete(n):
    return 3 * ((1 / 2) ** n) * U(n)


# Definición del paso de tiempo para tiempo continuo
dt = 0.001

# Definición del rango de tiempo continuo
t = np.arange(-5, 5 + dt, dt)

# Definición del rango de tiempo discreto
n = np.arange(-5, 6)

# Cálculo de las señales
y1 = exp_decay(t)
y2 = exp_decay_discrete(n)

# Gráfica de las señales
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, y1)
plt.title("Funcion de tiempo continuo t")
plt.grid(True)
plt.subplot(2, 1, 2)
plt.stem(n, y2, linefmt="b-", markerfmt="bo", basefmt=" ")
plt.title("Funcion de tiempo discreto n")
plt.grid(True)
plt.tight_layout()
plt.show()

# Cálculo de energías
Energia_cont = np.sum(np.abs(y1) ** 2) * dt
Energia_disc = np.sum(np.abs(y2) ** 2)
print("Energia de la senal continua:", Energia_cont)
print("Energia de la senal discreta:", Energia_disc)
```

# Definición de señales de entrada:

A continuación, el estudiante debe desarrollar dos señales continuas, $f(t)$ y $g(t)$ definidas a continuacíon:

$$
f(t) = \left\{ \begin{array}{cl}
1 & si \ 0 \le \ t \le 2\\
0 & en \ otro \ caso
\end{array} \right.
$$

$$
g(t) = \left\{ \begin{array}{cl}
t & si \ 0 \le \ t \le 1\\
0 & en \ otro \ caso
\end{array} \right.
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definición del rango de tiempo continuo
t = np.arange(0, 5, 0.01)


# Definición de la función f(t)
def f(t):
    val_f = []
    for x in t:
        if x >= 0 and x <= 2:
            val_f.append(1)
        else:
            val_f.append(0)
    return val_f


# Definición de la función g(t)
def g(t):
    val_g = []
    for x in t:
        if x >= 0 and x <= 1:
            val_g.append(x)
        else:
            val_g.append(0)
    return val_g


# Gráfica de las señales
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.title("Señal f(t)")
plt.plot(t, f(t))
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.subplot(2, 1, 2)
plt.title("Señal g(t)")
plt.plot(t, g(t))
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.tight_layout()
plt.grid(True)
plt.show()
```

# Convolución de señales continuas


```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01
t = np.arange(0, 5, dt)


# Definición de la convolución
def convolution(f, g, dt):
    conv = 0
    N = len(f)
    M = len(g)
    conv = np.zeros(N + M - 1)
    for n in range(N + M - 1):
        for k in range(max(0, n - M + 1), min(n + 1, N)):
            conv[n] += f[k] * g[n - k] * dt
    return conv


conv_res = convolution(f(t), g(t), dt)
time = np.arange(0, len(conv_res) * dt, dt)

plt.figure(figsize=(15, 2))
plt.plot(time, conv_res)
plt.title("Convolución de f(t) y g(t)")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

# Convolución señales discretas

Sean las siguiente señal discreta, calcular la convolución y graficar:
$$ x_1[n] = u[n] - u[n-3] $$

Calcular: $y[n] = x_1[n] * x_1[n]$


```python
import numpy as np
import matplotlib.pyplot as plt

# Vector discreto
n = np.arange(-2, 7)

# Señales a convolucionar
x1 = np.heaviside(n, 1) - np.heaviside(n - 3, 1)
x2 = x1.copy()

# Convolucion utilizando numpy
y = np.convolve(x1, x2)

# Nuevo vector temporal mas largo
nc = np.arange(-4, 13)

# Gráfica de las señales y la convolucion
plt.subplot(3, 1, 1)
# plt.stem(n, x1, 'ro')
plt.stem(n, x1, linefmt="b-", markerfmt="bo", basefmt=" ")
plt.grid(True)
plt.title("$x_1[n]$")

plt.subplot(3, 1, 2)
# plt.stem(n, x2, 'ro')
plt.stem(n, x2, linefmt="b-", markerfmt="bo", basefmt=" ")
plt.grid(True)
plt.title("$x_2[n] = x_1[n]$")

plt.subplot(3, 1, 3)
# plt.stem(nc, y ,'bx')
plt.stem(nc, y, linefmt="b-", markerfmt="bo", basefmt=" ")
plt.grid(True)
plt.title("$y[n]$")
plt.xlim([-2, 6])
plt.tight_layout()
plt.show()
```

### Actividad 3:

Obtener la energía de la señales anteriores: Señal Total, Señal sin ruido y Señal con ruido. Realizar una comparación de los resultados


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read

fs_señal_total, señal_total = read("../lab0/nuevo_audio.wav")
fs_señal_ruido, señal_ruido = read("../lab0/ruido.wav")
fs_señal, señal = read("../lab0/principito.wav")

dt = 0.001

energia_total = np.sum(np.abs(señal_total) ** 2) * dt
energia_ruido = np.sum(np.abs(señal_ruido) ** 2) * dt
energia_señal = np.sum(np.abs(señal) ** 2) * dt

print(f"Energia de la señal total: {round(energia_total)} Joules")
print(f"Energia de la señal de ruido: {round(energia_ruido)} Joules")
print(f"Energia de la señal: {round(energia_señal)} Joules")
```

**Conclusión**: La presencia de ruido en una señal puede tener un efecto significativo en la energía total de la señal combinada. Debido a la naturaleza aleatoria del ruido, los valores positivos y negativos del ruido pueden interactuar con los componentes de la señal original de tal manera que parcialmente se cancelen entre sí. Esto puede resultar en una reducción de la energía total de la señal combinada en comparación con la suma de las energías individuales de la señal con ruido y la señal sin ruido. En resumen, la adición de ruido a una señal puede disminuir su energía total debido a la cancelación parcial entre los componentes de señal y ruido.
